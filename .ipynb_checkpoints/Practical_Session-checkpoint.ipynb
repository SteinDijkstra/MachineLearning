```python
n=100000
center0=(-4.5,4.25)
center1=( 1.5,   1.75)
# center1=( -3,   3)
X,y,centers=make_blobs(n_samples=[n,n], n_features=2, centers=(center0,center1), 
                       cluster_std=1.0, shuffle=True,center_box=(-5.0, 5.0), 
                       return_centers=True,random_state=8112022)
```

```python
dist0 = multivariate_normal(mean=centers[0], cov=1)
dist1= multivariate_normal(mean=centers[1], cov=1)
```

```python
eta=(dist1.pdf(X)/(dist0.pdf(X)+dist1.pdf(X)))
```

```python
result= np.mean(np.minimum(eta,1-eta))
result
```

Q5

```python
def plotDecisionboundary(Xsample,ysample,clf):
    clf.fit(Xsample, ysample)

    plt.scatter(Xsample[:, 0], Xsample[:, 1], c=ysample, s=30, cmap=plt.cm.Paired)

    # plot the decision function
    ax = plt.gca()
    DecisionBoundaryDisplay.from_estimator(
        clf,
        Xsample,
        plot_method="contour",
        colors="k",
        levels=[-1, 0, 1],
        alpha=0.5,
        linestyles=["--", "-", "--"],
        ax=ax,
    )
    # plot support vectors
    ax.scatter(
        clf.support_vectors_[:, 0],
        clf.support_vectors_[:, 1],
        s=100,
        linewidth=1,
        facecolors="none",
        edgecolors="k",
    )
    plt.show()
```

```python
plotDecisionboundary(Xsample,ysample,10)
plotDecisionboundary(Xsample,ysample,1)
plotDecisionboundary(Xsample,ysample,0.1)
plotDecisionboundary(Xsample,ysample,0.01)
plotDecisionboundary(Xsample,ysample,0.001)
```

Q6

```python
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
n=1000
center0=(-1,1)
center1=(1,-1)
# center1=( -3,   3)
X,y,centers=make_blobs(n_samples=[n,n], n_features=2, centers=(center0,center1), 
                       cluster_std=1.0, shuffle=True,center_box=(-5.0, 5.0), 
                       return_centers=True,random_state=8112022)
```

```python
Coptions=np.logspace(-3, 1, num=20)
parameters={'kernel':['linear'],'C':Coptions}
svc = svm.SVC()

gridcv=GridSearchCV(svc, parameters,
                scoring='accuracy', refit=True, 
                cv=5, verbose=1,  return_train_score=True)

gridcv.fit(X,y)
```

```python
df=pd.DataFrame.from_dict(gridcv.cv_results_)
df[['param_C','rank_test_score','mean_test_score','std_test_score','mean_train_score', 'std_train_score']]
```

```python
plotDecisionboundary(X,y,gridcv.best_estimator_)
```

Q7

```python
n=200
center0=(0,2)
center1=(0,-1)
# center1=( -3,   3)
X,y,centers=make_blobs(n_samples=[n,n], n_features=2, centers=(center0,center1), 
                       cluster_std=[[3.0,1],[2.0,1]], shuffle=True,center_box=(-5.0, 5.0), 
                       return_centers=True,random_state=8112022)
```

```python
Coptions=np.logspace(-2, 1, num=10)
gammaOptions=np.logspace(-2, 1, num=10)
parameters={'kernel':['sigmoid'],'C':Coptions,"gamma":gammaOptions}
svc = svm.SVC()

gridcv=GridSearchCV(svc, parameters,
                scoring='accuracy', refit=True, 
                cv=5, verbose=1,  return_train_score=True)

gridcv.fit(X,y)
```

```python
df=pd.DataFrame.from_dict(gridcv.cv_results_)
res=df[['param_C','param_gamma','rank_test_score','mean_test_score','mean_train_score','params']]
res=res.sort_values("mean_train_score",ascending=False)
res.loc[:,res.columns!="params"][:10]
```

```python
gridcv.best_params_
```

```python
c=res.loc[:,"params"][0]["C"]
gamma=res.loc[:,"params"][0]["gamma"]
plotDecisionboundary(X,y,svm.SVC(C=c,kernel="sigmoid",gamma=gamma))
```

```python
plotDecisionboundary(X,y,gridcv.best_estimator_)
```
